# Ex1 - Getting and Knowing your Data


### Step 1: Initialize PySpark Session



```python
from pyspark.sql import SparkSession

# Create a Spark session
spark = SparkSession.builder.appName("Day1").getOrCreate()
```

### Step 2: Load the Dataset





```python
# Load the Chipotle dataset into a Spark DataFrame
data_path = '/home/riyaz/Downloads/chipotle.csv' # Replace with the actual path
df = spark.read.csv(data_path, header=True, inferSchema=True)
```

### Step 3. Get an overview of the DataFrame:


```python
df.printSchema()
```

```python
df.show(5)
```

### Step 4.Calculate basic statistics:


```python
df.describe().show()
```

```python
df.summary().show()
```

### Step 5. What is the number of observations in the dataset?

```python
print(f"Total number of observations are: {df.count()}")
```

### Step 6. What is the number of columns in the dataset?

```python
print(f"Total number of columns in the data set are: {len(df.columns)}")
```

### Step 7. Print the name of all the columns.

```python
print(f"Names of the columns in the dataset are: {df.columns}")
```
